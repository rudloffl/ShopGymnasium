```python
# CELLULE 2 — Fonction d’affichage d’un état

def print_obs(obs, prefix="OBS"):
    print(
        f"{prefix} | t={int(obs[0])} | "
        f"M1(busy={int(obs[1])}, left={int(obs[2])}) | "
        f"M2(busy={int(obs[3])}, left={int(obs[4])}) | "
        f"RAW={int(obs[5])} | P1={int(obs[6])} | P2inter={int(obs[7])} | P2={int(obs[8])} | "
        f"next_deliv={int(obs[9])} | backlogP1={int(obs[10])} | backlogP2={int(obs[11])} | "
        f"en_route={int(obs[12])}"
    )
```

```python
# CELLULE 3 — Reset + affichage initial
env = WorkshopEnv()
obs, info = env.reset()

print("=== RESET ===")
print_obs(obs)
```

```python
print("===== TEST LIVRAISONS MULTIPLES (avec jitter ±2) =====")

env.reset()

def print_obs(obs):
    print(f"OBS | t={int(obs[0])} | RAW={int(obs[5])} "
          f"| next_deliv={int(obs[9])} | en_route={int(obs[12])} "
          f"| p1={int(obs[6])} | p2={int(obs[8])}")

# --------------------------------------------
# 1) Envoyer 3 commandes MP
# --------------------------------------------
times_commands = []
for i in range(3):
    obs, r, *_ = env.step(150)   # action = commander MP (q = 1)
    times_commands.append(env.time)
    print(f"\nCommande MP #{i+1} envoyée à t={env.time}, reward={r}")
    print_obs(obs)

# --------------------------------------------
# 2) Avancer jusqu'à ce que TOUTES les livraisons soient arrivées
# --------------------------------------------
deliveries_detected = []
previous_raw = int(obs[5])

while env.time < 300:  # marge confortable
    obs, r, *_ = env.step(200)

    current_raw = int(obs[5])

    # Détection livraison : RAW augmente
    if current_raw > previous_raw:
        delivered = current_raw - previous_raw
        deliveries_detected.append((env.time, delivered))
        print(f"\n>>> LIVRAISON détectée à t={env.time} : "
              f"+{delivered} MP (RAW={current_raw})")
    previous_raw = current_raw

# --------------------------------------------
# 3) Résumé final
# --------------------------------------------
print("\n===== RÉSUMÉ DES LIVRAISONS =====")
for i, (t_cmd) in enumerate(times_commands):
    print(f"Commande #{i+1} envoyée à t={t_cmd}")

print("\nLivraisons arrivées :")
for arr in deliveries_detected:
    print(f"- t={arr[0]} → livraison de {arr[1]} MP")

print("\n===== ANALYSE ATTENDUE =====")
print("Chaque livraison doit arriver dans l'intervalle [t_cmd + 118, t_cmd + 122].")
print("Aucune livraison ne doit écraser une autre.")
print("En_route doit décroître jusqu'à 0.")
```

```python
print("===== TEST LIVRAISON + VENTE DANS LA MÊME MINUTE =====")

env.reset()

def print_obs(obs):
    print(f"OBS | t={int(obs[0])} | RAW={int(obs[5])} "
          f"| P1={int(obs[6])} | P2={int(obs[8])} "
          f"| backlogP1={int(obs[10])} | backlogP2={int(obs[11])} "
          f"| next_deliv={int(obs[9])} | en_route={int(obs[12])}")

# --------------------------------------------------
# 1) Préparation : on produit artificiellement du stock à vendre
# --------------------------------------------------
env.stock.p1 = 10
env.stock.p2 = 5

# --------------------------------------------------
# 2) On envoie une commande MP à t = 0
# --------------------------------------------------
obs, r, *_ = env.step(150)  # commande q = 1
print(f"\nCommande envoyée à t={env.time}, en_route={int(obs[12])}")
print_obs(obs)

# --------------------------------------------------
# 3) Avancer jusqu'à t = 115 pour s'approcher de la zone critique
# --------------------------------------------------
while env.time < 115:
    obs, r, *_ = env.step(200)

print("\n--- Arrivée proche de t = 120 ---")
print_obs(obs)

# --------------------------------------------------
# 4) Avancer minute par minute et surveiller :
#    - ventes à t % 60 == 0
#    - livraison (augmentation RAW)
# --------------------------------------------------
previous_raw = int(obs[5])
previous_p1 = int(obs[6])
previous_p2 = int(obs[8])
previous_backlog1 = int(obs[10])

for _ in range(10):  # couvrira t = 116 → 125
    obs, r, *_ = env.step(200)
    t = int(obs[0])
    raw = int(obs[5])
    p1 = int(obs[6])
    p2 = int(obs[8])

    # Détection livraison
    if raw > previous_raw:
        print(f"\n>>> LIVRAISON détectée à t={t} : RAW passe de {previous_raw} à {raw}")
    
    # Détection vente
    if t % 60 == 0:
        print(f"\n>>> VENTE AUTOMATIQUE À t={t}")
        print(f"Stock après vente : P1={p1}, P2={p2}")
        print(f"Backlog après vente : backlogP1={int(obs[10])}, backlogP2={int(obs[11])}")

    previous_raw = raw

print("\n===== FIN TEST =====")
print_obs(obs)

print("\n===== ANALYSE ATTENDUE =====")
print("- Une livraison devrait tomber entre t = 118 et 122.")
print("- Si elle tombe EXACTEMENT à t = 120, elle sera simultanée à une vente.")
print("- La vente ne doit JAMAIS consommer les MP (RAW).")
print("- La livraison n'affecte JAMAIS P1, P2 ou leurs backlogs.")
```

```python
print("===== TEST : LIVRAISON + VENTE + VOL NOCTURNE =====")

env.reset()

def print_obs(label, obs):
    print(f"\n--- {label} (t={int(obs[0])}) ---")
    print(f"RAW={int(obs[5])} | P1={int(obs[6])} | P2={int(obs[8])} | "
          f"backlogP1={int(obs[10])} | backlogP2={int(obs[11])} | "
          f"next_deliv={int(obs[9])} | en_route={int(obs[12])}")

# --------------------------
# 1. Préparation de l'état
# --------------------------
env.stock.p1 = 20
env.stock.p2 = 15
env.market.backlog_p1 = 30
env.market.backlog_p2 = 20

# Une commande envoyée tard → livraison proche de minuit
# Ex : en envoyant à t = 1315, on aura une livraison vers 1433-1437
t_target_cmd = 1315

# Avancer à t = 1315
while env.time < t_target_cmd:
    env.step(200)

# Envoyer commande MP
obs, r, *_ = env.step(150)  # q = 1
print_obs("Commande MP envoyée", obs)

# --------------------------
# 2. Avancer minute par minute
#    vers la zone critique :
#    1420 → 1440
# --------------------------
previous_raw = int(obs[5])
previous_p1 = int(obs[6])
previous_p2 = int(obs[8])

events = []

for _ in range(150):  # couvre ~1316 → 1465
    obs, r, *_ = env.step(200)
    t = int(obs[0])
    raw = int(obs[5])
    p1 = int(obs[6])
    p2 = int(obs[8])

    # LIVRAISON
    if raw > previous_raw:
        events.append(f"LIVRAISON à t={t} (RAW {previous_raw} → {raw})")

    # VENTE
    if t % 60 == 0:
        events.append(f"VENTE à t={t} (P1={p1}, P2={p2}, backlog={int(obs[10])}/{int(obs[11])})")

    # VOL NOCTURNE
    if t == env.theft_time:

        events.append(f"VOL NOCTURNE à t={t} (P1={previous_p1} → {p1}, P2={previous_p2} → {p2})")

    previous_raw = raw
    previous_p1 = p1
    previous_p2 = p2

print_obs("État final", obs)

print("\n===== ÉVÉNEMENTS DÉTECTÉS =====")
for e in events:
    print(e)

print("\n===== ANALYSE ATTENDUE =====")
print("- La livraison doit arriver vers t ≈ 1433–1437.")
print("- La vente la plus proche est à t = 1380, puis la suivante à t = 1440.")
print("- Le vol nocturne se produit à t = 1435.")
print("- Aucun événement ne doit écraser un autre.")
print("- Le vol ne doit jamais affecter RAW.")
print("- Une livraison proche de t=1435 ne doit pas perturber le vol.")
print("- La vente à t=1440 doit ignorer totalement le vol.")
```
